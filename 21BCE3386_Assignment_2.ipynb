## **Task 3:**
 Perform the Below Visualizations.

● Univariate Analysis 

● Bi - Variate Analysis 

● Multivariate Analysis

## Univariate Analysis

````python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


plt.figure(figsize=(8, 6))
sns.histplot(df['number of bedrooms'], bins=20, kde=True)
plt.title('Distribution of Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Frequency')
plt.show()
````

````python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['number of bedrooms'])
plt.title('Boxplot of Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.show()
````

````python
plt.figure(figsize=(8, 6))
sns.countplot(x=df['condition of the house'])
plt.title('Countplot of Condition of the House')
plt.xlabel('Condition of the House')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
````

## Bi - Variate Analysis

````python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='living area', y='Price', data=df)
plt.title('Scatter Plot of Living Area vs. Price')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
````

````python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
````

````python
plt.figure(figsize=(10, 6))
sns.boxplot(x='number of bedrooms', y='Price', data=df)
plt.title('Box Plot of Number of Bedrooms vs. Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
````

## Multivariate Analysis

````python
pairplot = sns.pairplot(df, vars=['living area', 'Price'], hue='number of bathrooms')
pairplot.fig.suptitle('Pairplot of Living Area, Price, and Number of Bathrooms', y=1.02)
plt.show()
````

````python
multivariate_corr_matrix = df[['living area', 'number of bedrooms', 'number of bathrooms', 'Price']].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(multivariate_corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Multivariate Correlation Matrix')
plt.show()
````
