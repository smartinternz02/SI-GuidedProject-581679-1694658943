##### Reducing Imbalance in data by splitting into good and bad category

````python
indexBad = df.loc[df['quality'] <= 6].index
indexGood = df.loc[df['quality'] > 6].index
df.iloc[indexBad, df.columns.get_loc('quality')] = 0
df.iloc[indexGood, df.columns.get_loc('quality')] = 1
````

````python
df['quality'].value_counts()
````

````python
fig = plt.figure(figsize = (3,2))
sns.set(style="whitegrid")
sns.countplot(x=df['quality'])
plt.show()
````

As we can see data is highly imbalanced

##### OUTLIER REMOVAL
##### For Skewed Columns lets do IQR Outlier removal

````python
def replace_outliers_with_quartiles(column):
    # Calculate the quartiles and IQR for the column
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    
    # Calculate the lower and upper bounds for outliers
    fence_low = q1 - 1.5 * iqr
    fence_high = q3 + 1.5 * iqr
    
    # Replace outliers with quartile boundaries
    column[column < fence_low] = q1
    column[column > fence_high] = q3
    
    return column

df[skewed_cols] = df[skewed_cols].apply(replace_outliers_with_quartiles)
````

````python
fig = plt.figure(figsize = (8,4))
plt.xticks(rotation = 45)
sns.set(style="whitegrid")
sns.boxplot(data=df[skewed_cols])
plt.show()
````

## **Task 3:**
Machine Learning Model building

````python
y = df.quality
X = df.drop('quality', axis=1)
````

````python
#### REMOVING IMBALANCE using SMOTE
# from imblearn.over_sampling import SMOTE
# oversample = SMOTE(k_neighbors=4)
# X, y = oversample.fit_resample(X, y)

### GENERATED BIAS, hence not using
````

````python
y.value_counts()
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=45)
````

````python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
````

## **Task 4:**
Evaluate the model

````python
from sklearn.metrics import accuracy_score

y_predict = model.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)*100
accuracy
````

````python
from sklearn.metrics import f1_score

f1=f1_score(y_test,y_predict,average='weighted')
f1
````

````python
from sklearn.metrics import confusion_matrix

conf_mat = confusion_matrix(y_test, y_predict)
print(conf_mat)
````

````python
from sklearn.metrics import classification_report

class_report = classification_report(y_test, y_predict)
print(class_report)
````
